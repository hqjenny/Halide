```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
costs = []

i = 0
while True:
    try:
        path = '../random_pipeline/results/costs_{}.txt'.format(i)
        with open(path, 'r') as file:
            print(path)
            for line in file:
                costs.append(float(line.rstrip()))
        i += 1
    except FileNotFoundError:
        break
```

```python
runtimes = []

i = 0
while True:
    try:
        path = '../random_pipeline/results/runtimes_{}.txt'.format(i)
        with open(path, 'r') as file:
            print(path)
            for line in file:
                runtimes.append(float(line.rstrip()))
        i += 1
    except FileNotFoundError:
        break
```

```python
len(costs)
```

```python
len(runtimes)
```

```python
# Trained with random pipeline generator with faster learning rate
plt.xlabel("Cost")
plt.ylabel("Runtime")
plt.scatter(costs, runtimes)
print("Correlation:", np.corrcoef(costs, runtimes)[0, 1])
```
